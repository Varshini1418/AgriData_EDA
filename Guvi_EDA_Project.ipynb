## DATA CLEANSING

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\VARSHINI\Desktop\Guvi Classes\ICRISAT-District Level Data - ICRISAT-District Level Data.csv")
print(df.info())
print(df.head())
```

```python
# Strip whitespace and convert to lowercase with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("(", "").str.replace(")", "")
```

```python
df.head()
```

```python
# Check missing values
print(df.isnull().sum())
```

```python
#Filling crop yields with mean where missing
df['rice_yield_kg_per_ha'] = df['rice_yield_kg_per_ha'].fillna(df['rice_yield_kg_per_ha'].mean())
df['wheat_yield_kg_per_ha'] = df['wheat_yield_kg_per_ha'].fillna(df['wheat_yield_kg_per_ha'].median())

# Drop rows with missing district or year info
df = df.dropna(subset=['dist_name', 'year'])
```

# EDA

```python
import plotly.express as px
top_rice = df.groupby("state_name")["rice_production_1000_tons"]\
    .sum().sort_values(ascending=False).head(7).reset_index()

px.bar(
    top_rice,
    x="state_name",
    y="rice_production_1000_tons",
    title="Top 7 Rice Producing States in India",
    color="state_name",
    text="rice_production_1000_tons"
).show()
```

```python
top_wheat = df.groupby("state_name")["wheat_production_1000_tons"]\
    .sum().sort_values(ascending=False).head(5).reset_index()
total = top_wheat["wheat_production_1000_tons"].sum()
top_wheat["percentage_share"] = (top_wheat["wheat_production_1000_tons"] / total) * 100

# Bar chart
px.bar(
    top_wheat,
    x="state_name",
    y="wheat_production_1000_tons",
    title="Top 5 Wheat Producing States",
    color="state_name",
    text="wheat_production_1000_tons"
).show()

# Pie chart
px.pie(
    top_wheat,
    names="state_name",
    values="percentage_share",
    title="Percentage Share of Wheat Production",
    
).show()
```

```python
top_oilseed=df.groupby("state_name")["oilseeds_production_1000_tons"]\
    .sum().sort_values(ascending=False).head(5).reset_index()

px.histogram(
    top_oilseed,
    x="state_name",
    y="oilseeds_production_1000_tons",
    title="Top 5 Oilseeds Producing States in India",
    color="state_name"
).show()
```

```python
top_sunflower=df.groupby("state_name")["sunflower_production_1000_tons"]\
.sum().sort_values(ascending=False).head(7).reset_index()

px.line(
    top_sunflower,
x="state_name",
y="sunflower_production_1000_tons",
title="Top 7 Sunflower Producing States in India",
color="state_name",
markers=True
).show()
```

```python
ly = df["year"].max()
cy = ly - 50

# Filter last 50 years
recent_df = df[df["year"] >= cy]

# Group and summarize sugar production
top_sugar = recent_df.groupby("year")["sugarcane_production_1000_tons"]\
    .sum().reset_index()

# Plot the line chart
px.line(
    top_sugar,
    x="year",
    y="sugarcane_production_1000_tons",
    title=f"Sugarcane Production in India (Last 50 Years: {cy}â€“{ly})",
    markers=True
).show()

```

```python
import plotly.express as px
ly = df["year"].max()
cy = ly - 50

# Filter last 50 years
recent_df = df[df["year"] >= cy]

# Group and summarize production data
trend = recent_df.groupby("year")[["rice_production_1000_tons", "wheat_production_1000_tons"]]\
    .sum().reset_index()

# Plot the line chart with proper title formatting
px.line(
    trend,
    x="year",
    y=["rice_production_1000_tons", "wheat_production_1000_tons"],
    title=f"Rice Production Vs Wheat Production (Last 50 Years: {cy}â€“{ly})",
    markers=True
).show()
```

```python
import plotly.graph_objects as go

# Group and prepare the data
soyabean = df.groupby('state_name')[['soyabean_production_1000_tons', 'soyabean_yield_kg_per_ha']]\
    .sum().sort_values(by="soyabean_production_1000_tons", ascending=False).head(5).reset_index()

# Create the figure
fig = go.Figure()

# Add bar chart for production
fig.add_trace(go.Bar(
    x=soyabean['state_name'],
    y=soyabean['soyabean_production_1000_tons'],
    name='Production (1000 tons)',
    marker_color='steelblue'
))

# Add line chart for yield
fig.add_trace(go.Scatter(
    x=soyabean['state_name'],
    y=soyabean['soyabean_yield_kg_per_ha'],
    name='Yield (kg/ha)',
    yaxis='y2',
    mode='lines+markers',
    line=dict(color='darkorange', width=2),
    marker=dict(size=7)
))

# Layout with dual y-axes
fig.update_layout(
    title='Soybean Production vs Yield Efficiency (Top 5 States)',
    xaxis_title='State Name',
    yaxis=dict(title='Production (1000 tons)', side='left'),
    yaxis2=dict(title='Yield Efficiency (kg/ha)', overlaying='y', side='right'),
    legend=dict(x=0.05, y=1.1, orientation='h'),
    template='plotly_white',
    width=800,
    height=500
)

fig.show()
```
